# 随机漫步

```python
import random

import matplotlib.pyplot as plt
import numpy as np

position = 0
walk = [position]
steps = 10000
for i in range(steps):
    step = 1 if random.randint(0,1) else -1
    position += step
    walk.append(position)

plt.plot(walk[:100])
```

```python
nsteps = 1000
draws = np.random.randint(0,2,size=nsteps)
steps = np.where(draws > 0, 1, -1)
walk = steps.cumsum()
walk.min()
```
